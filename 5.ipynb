```python
import numpy as np
import matplotlib.pyplot as plt

n = 20  
xbar = 30   
s2 = 20 

a = 1      
b = 1      
c = 3      
d = 1 / 100

N = 100000          # Number of iterations

mu_samples = np.zeros(N)
tau_samples = np.zeros(N)

mu_current = a / c      # mu = a / c
tau_current = c         # tau = c

for i in range(N):
    muprec = n * tau_current + d
    mu_mean = (c * d + n * tau_current * xbar) / muprec
    mu_var = 1 / muprec
    mu_current = np.random.normal(mu_mean, np.sqrt(mu_var))
    
    alpha_tau = a + n / 2
    beta_tau = b + 0.5 * ((n - 1) * s2 + n * (xbar - mu_current) ** 2)
    tau_current = np.random.gamma(shape=alpha_tau, scale=1 / beta_tau)
    
    mu_samples[i] = mu_current
    tau_samples[i] = tau_current
```

### (i)

```python
plt.figure(figsize=(10, 4))
plt.plot(mu_samples, color='blue')
plt.title('Trace Plot of μ')
plt.xlabel('Iteration')
plt.ylabel('μ')
plt.show()
```

### (ii)

```python
plt.figure(figsize=(8, 6))
plt.hist(mu_samples, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Marginal Distribution of μ')
plt.xlabel('μ')
plt.ylabel('Density')
plt.show()
```

### (iii)

```python
plt.figure(figsize=(10, 4))
plt.plot(tau_samples, color='red')
plt.title('Trace Plot of τ')
plt.xlabel('Iteration')
plt.ylabel('τ')
plt.show()
```

### (iv)

```python
plt.figure(figsize=(8, 6))
plt.hist(tau_samples, bins=50, density=True, color='salmon', edgecolor='black')
plt.title('Marginal Distribution of τ')
plt.xlabel('τ')
plt.ylabel('Density')
plt.show()
```

### (v)

```python
plt.figure(figsize=(8, 6))
plt.scatter(mu_samples, tau_samples, s=1, alpha=0.5)
plt.title('Joint Posterior Distribution of μ and τ')
plt.xlabel('μ')
plt.ylabel('τ')
plt.show()
```

### (vi)

```python
mu_mean = np.mean(mu_samples)
print(f"Mean of mu: {mu_mean}")
```

### (vii)

```python
tau_mean = np.mean(tau_samples)
print(f"Mean of tau: {tau_mean}")
```
